## Part 2

(11)

```python
np.random.seed(42)
a_array = np.random.randint(1, 10, size = (5,4))
A = np.matrix(a_array)
A
```

(12)

```python
B = np.dot(A.transpose(), A)
B
```

(13)

```python
B.transpose() - B
# since the result = 0, B is a symmetric matrix.
```

(14)

```python
eig_values = np.linalg.eig(B)[0]
# eigen values
```

```python
eig_vectors = np.linalg.eig(B)[1]
# eigenvectors
eig_vectors
```

(15)

```python
e = 1e-10
```

```python
for i in range(len(eig_vectors)):
    for j in range(i + 1, len(eig_vectors)):
        dp = np.dot(eig_vectors[:,i].transpose(),eig_vectors[:,j])
        if np.abs(dp) > e:
            print(f"{eig_vectors[:,i].transpose()} and {eig_vectors[:,j]} are not orthogonal")
        else:
            print("pass!")
```

(16)

```python
np.linalg.det(B)
```

(17)

```python
result = 1
for value in eig_values:
    result *= value
result
# similar to result in (16)
```
